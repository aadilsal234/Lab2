```python
try:
    with open("genome.txt", "r") as file:
        genome_sequence = file.read()
except FileNotFoundError:
    print("Error")
    genome_sequence = ""

gen_list = list(genome_sequence)
print(gen_list)
gen_length = len(gen_list)
print(gen_length)

t=np.linspace(0,4*np.pi,gen_length)
x=np.cos(t)
y=np.sin(t)
z=np.linspace(0,5,gen_length)

cooridnates=np.column_stack((x,y,z))
#print(cooridnates)

fig=plt.figure()
ax=plt.axes(projection="3d")
colors={'A':'red','T':'blue','C':'green','G':'yellow'}
for i in range(gen_length):
        ax.scatter3D(cooridnates[i,0],cooridnates[i,1],cooridnates[i,2],color=colors[gen_list[i]])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Genome Sequence")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO


image_url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTepI0o3nI1baWjVjlTdLkFl7Nx1uoZJRS80w&s"
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

image_np = np.array(img)

plt.figure(figsize=(8, 6))
plt.imshow(image_np)
plt.title("Og Image")
plt.show()

rotated_image = np.rot90(image_np)
plt.figure(figsize=(8, 6))
plt.imshow(rotated_image)
plt.title("Rotated Image")
plt.show()

flipped_image = np.fliplr(image_np)
plt.figure(figsize=(8, 6))
plt.imshow(flipped_image)
plt.title("Flipped Image")
plt.show()

grayscale_image = np.dot(image_np[..., :3], [0.2989, 0.5870, 0.1140])
plt.figure(figsize=(8, 6))
plt.imshow(grayscale_image, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np

iris=load_iris()
X=np.array(iris.data)
Y=np.array(iris.target)

print("Mean")
print(np.mean(X,axis=0))
print("Median")
print(np.median(X,axis=0))
print("Standard Deviation")
print(np.std(X,axis=0))
print("Minimum")
print(np.min(X,axis=0))
print("Maximum")
print(np.max(X,axis=0))

sepal_d = X[:, :2]
print("Sepal Data")
print(sepal_d)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Scatter Plot(Sepal Length vs. Sepal Width)")
plt.show()

plt.figure(figsize=(8,6))
plt.hist(X[:, 0], bins=10)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram(Distribution of Sepal Length)")
plt.show()

plt.figure(figsize=(8,6))
plt.plot(X[:,2],X[:,3])
plt.xlabel("Petal Length")
plt.ylabel("Petal width")
plt.title("Line Plot(PETAL LENGTH VS PETAL WIDTH)")
plt.show()

```
